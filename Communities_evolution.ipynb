# Communities evolution

At this point we have applied the different community detection algorithms for networks build for each year and for each commodity.  Now we want to compare the communites found in each year for a specific product for two reasons, first to check if the communities remain toghether along the time which would mean that the algorithm is detecting the same pattern and we can say that a community is really present. And the other reason is that we can study the communities evolutions, things like the number of nodes, the number of links and also other centrality measures.

But before doing this we have to solve a few problems. We want to compare communities detected at different years, how can we say that community A belonging to year 2001 is the same as community B belonging to 2002? Our proposal is to measure the simmilarity between the two sets where we defined the similarity as:

$$ sim(A,B) = \frac{A \cap B}{A \cup B}$$

this will output a value between 0 and 1 that we can use to measure how similar two communities are. 

So our aim in this notebook is to associate all the communities that have the higger simmilarity between years, we know it may fail with the smaller ones but we will keep track of the bigger communities and in this way we can estimate communitiy properties and compare them over time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
import geopandas as gpd
import math
import imageio
from IPython.display import Image
```

```python
def country_coordinates(path):
    '''Get the country coordinates to be used for plotting purpose'''
    c_fields     = ["cc3","latitude","longitude"]
    df_coor = pd.read_csv(path, skipinitialspace = True, usecols = c_fields)
    df_coor = df_coor.rename(columns={"cc3": "Node"})
    return(df_coor)
```

```python
def get_all_years(algorithm,commodity):
    '''This function get the communities obtained with 
       a specific     algorithm
       and a specific commodity
       for all time perior 2001 to 2015'''
    df = country_coordinates("../NS_FP_DATA/countries.csv")
    i = 3
    for year in range(2001,2021):
        df_read = pd.read_csv(f"Data/Community/{algorithm}_{year}_{commodity}_communities.csv") 
        df = pd.merge(df, df_read, on='Node', how='outer')
        df = df.rename(columns={df.columns[i]: year})
        i +=1
    df = df.dropna(subset=[2001, 2002,2003,2004,2005, 2006,2007,2008,2009,2010,
                           2011,2012,2013,2014,2015,2016,2017,2018,2019,2020], how='all')
    return(df)
```

```python
def com_sim(community_1,community_2):
    '''Quantify how similar two communities are
    community_i has to be a list of lists, 
    where each list has the community member isocode'''
    inter = np.intersect1d(community_1,community_2)
    union = np.union1d(community_1,community_2)
    return len(inter)/len(union)
```

```python
def assign_colors(labels_a,labels_b,ranks,thre_rank):
    '''This function transfer the labels of each community in a set a
       to the labels of the communities in a second set b'''
    _ranks = ranks.copy()
    row,col = np.unravel_index(np.argmax(_ranks,axis=None),_ranks.shape)
    maxi_rank = _ranks[row,col]
    # kinda inefficient to calculate the max value of a huge matrix each time
    while maxi_rank > thre_rank:
        labels_b[col] = labels_a[row]
        _ranks[row,:] = thre_rank
        _ranks[:,col] = thre_rank
        row,col = np.unravel_index(np.argmax(_ranks,axis=None),_ranks.shape)
        maxi_rank = _ranks[row,col]
```

```python
def community_maps_time(df_all,col,n=10):
    '''This function plot the n most important communites
       in a given year, the colors to use must be passed as a list'''
    
    fig, axs = plt.subplots(11,2, figsize=(14, 30), facecolor='w', edgecolor='k')
    #fig.subplots_adjust(hspace = .5, wspace=.001)

    axs = axs.ravel()
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    year = [i for i in range(2001,2021)]
    
    for i in range(20):
        world.boundary.plot(ax=axs[i])
    
        df_plot = df_all[["latitude","longitude",year[i]]].dropna()
        df_plot = df_plot[df_plot[year[i]] < n]
        key = [i for i in range(0,(int(max(df_all[year[i]].dropna().unique()))+1))]
        values = col[0:(int(max(df_all[year[i]].dropna().unique()))+1)]
        colors = dict(zip(key, values))
        axs[i].scatter(df_plot["longitude"],df_plot["latitude"], c=df_plot[year[i]].map(colors))

        #scatter1 = axs[i].scatter(df_communities["longitude"],df_communities["latitude"], c=df_communities[year[i]], cmap="tab20c")
        axs[i].set_title(f'Year: {year[i]}')
        axs[i].set_ylabel("Latitude")
        axs[i].set_xlabel("Longitude")
        axs[i].grid(True)
        # produce a legend with the unique colors from the scatter
    plt.tight_layout()
    plt.show()
```

```python
def communities_identification(df_all,tolerance = 0.2):
    
    years = [i for i in range(2001,2020)]

    for year_1 in years:
        year_2 = year_1+1
    
        #Define the communities
        df = df_all[["Node",year_1]]
        df = df.dropna()
        df =df.pivot(columns=year_1,values="Node")
        communities_a =[list(df[i].dropna()) for i in list(df.keys())]

        df = df_all[["Node",year_2]]
        df = df.dropna()
        df =df.pivot(columns=year_2,values="Node")
        communities_b =[list(df[i].dropna()) for i in list(df.keys())]
    
    
        ext_labels = np.arange(len(communities_b)+len(communities_a))
        ranks = np.array([[com_sim(a,b) \
                       for b in communities_b]\
                       for a in communities_a])
    
        labels_a = ext_labels[:len(communities_a)]
        labels_b = ext_labels[-len(communities_b):]
        assign_colors(labels_a,labels_b,ranks,tolerance)
        #Now we have to rename the codes corresponding to the seccond year according to the new labels
        #Arrange a dictionary 
        if len(labels_b)>len(labels_a):
            key =  np.append(labels_a,[i for i in range(len(labels_a),len(labels_b))])
        else:
            key =  labels_a
        a_dictionary = dict(zip(key, labels_b))
        df_all[year_2].replace(a_dictionary, inplace=True)

        #In case the community number increase one of the old big labels will remain
        #so i'll reduce it to have numbers in the same range
        if len(labels_b)>len(labels_a):
            key    = [x for x in labels_b if x>max(labels_a)]
            values = [max(labels_a)+i for i in range(1,len(labels_b)-len(labels_a)+1)]
            a_dictionary = dict(zip(key, values))
            df_all[year_2].replace(a_dictionary, inplace=True)
            
    return(df_all)
```

```python
def community_maps_time(df,n=5):
    fig, axs = plt.subplots(10,2, figsize=(14, 30), facecolor='w', edgecolor='k')
    axs = axs.ravel()

    year = [i for i in range(2001,2021)]
    for i in range(20): 
        col= ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']
        df_plot = df[["latitude","longitude",year[i]]].dropna()
        #df_plot = df_plot[df_plot[year[i]] < n]

        key = [i for i in range(0,(int(max(df[year[i]].dropna().unique()))+1))]
        values = col[0:(int(max(df[year[i]].dropna().unique()))+1)]
        colors = dict(zip(key, values))

        country = []
        for j in df[year[i]]:
            if math.isnan(j)==False:
                if j < n:
                    country.append(colors[j])
                else:
                    country.append('grey')
            else:
                country.append('grey')
        
        df.plot(ax=axs[i], color = country)
        axs[i].set_title(f'Year: {year[i]}')
        #axs[i].set_ylabel("Latitude")
        #axs[i].set_xlabel("Longitude")
        axs[i].set_yticklabels([])
        axs[i].set_xticklabels([])
    #    axs[i].grid(True)
    plt.tight_layout()
    plt.savefig('Figures/communities_in_time.png')
    plt.show()  
```

```python
#Get data for one commodity
commodity = 1
#upload data
df_all_years = get_all_years("Spinglass",commodity)
df_all = communities_identification(df_all_years,0.2)
```

```python
def community_maps_time_gif(df,gifname,n=5):
    year = [i for i in range(2001,2021)]
    
    gif_files = []
    for i in range(20):
        name = f'GIF/{year[i]}.png'
        gif_files.append(name)
        
        plt.figure(figsize=(12, 4), dpi=200)
        col= ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']
        df_plot = df[["latitude","longitude",year[i]]].dropna()
        #df_plot = df_plot[df_plot[year[i]] < n]

        key = [i for i in range(0,(int(max(df[year[i]].dropna().unique()))+1))]
        values = col[0:(int(max(df[year[i]].dropna().unique()))+1)]
        colors = dict(zip(key, values))

        country = []
        for j in df[year[i]]:
            if math.isnan(j)==False:
                if j < n:
                    country.append(colors[j])
                else:
                    country.append('grey')
            else:
                country.append('grey')
        
        df.plot(color = country)
        plt.title(f'Year: {year[i]}')
        plt.ylabel("Latitude")
        plt.xlabel("Longitude")
        
    #    axs[i].grid(True)
        plt.savefig(name)
    plt.show()
    
    # Build GIF
    with imageio.get_writer(f'{gifname}.gif', mode='I') as writer:
        for filename in gif_files:
            image = imageio.imread(filename)
            writer.append_data(image)
    gif = imageio.mimread(f'{gifname}.gif')
    gif_speed_up = f'{gifname}.gif'
    imageio.mimsave(gif_speed_up, gif, fps=2)
```
